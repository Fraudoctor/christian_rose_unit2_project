#### Calculate the average star rating for each genre, but only include genres with at least 10 movies


**You only need to choose one of the options below.** However, solving for multiple options would count as "extra credit" that could e.g. raise your score from a 2 ("meets expectations") to a 3 ("exceeds expectations").

#### Option 1: manually create a list of relevant genres, then filter using that list

```python
# Answer:
movies.loc[:,'genre'].value_counts()#Counts of the genres
genre_greater_than_ten = movies.loc[:,'genre'].isin(['Drama', 'Comedy', 'Action', 'Crime','Biography','Adventure','Animation','Horror','Mystery']) #Manually selecting the ones that are over 10
movies.loc[genre_greater_than_ten , :] #Movies in those genres
movies.loc[genre_greater_than_ten , :].groupby('genre')['star_rating'].mean() #The mean rating for the above genres.
```

#### Option 2: automatically create a list of relevant genres by saving the value_counts and then filtering

```python
# Answer:
genres = movies['genre'].value_counts()[movies['genre'].value_counts() > 10].index #Create a list of movies that satisfy a criteria
movies[movies['genre'].isin(genres)].groupby('genre')['star_rating'].mean() #Apply this criteria to the original to then get the star rating for the movies in genres that satisfy our query.
```

#### Option 3: calculate the average star rating for all genres, then filter using a boolean Series

```python
# Answer:
movie_means = movies.groupby('genre')['star_rating'].mean() # Average Star rating for all genres.
movie_means[movies.loc[:,'genre'].value_counts()>=10] #Here we pass the boolean series into the subset. This only works because both are the same length. Though I am not sure how they know where to align or what ordering system is used for each row of each column.
```

#### Option 4: aggregate by count and mean, then filter using the count

```python
# Answer:
movie_genres = movies.groupby('genre')['star_rating'].agg(['count', 'mean'])
movie_genres.loc[movie_genres.loc[:,'count']>=10,:]
```
